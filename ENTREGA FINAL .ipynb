## Explicação da hipótese e mecanismo

A OTAN é uma aliança de 28 países na fronteira com o Oceano Atlântico Norte. Inclui os Estados Unidos, a maioria dos membros da União Europeia, o Canadá e a Turquia. A OTAN é um acrônimo para a Organização do Tratado do Atlântico Norte. A missão da OTAN é proteger a liberdade de seus membros. Seus alvos incluem armas de destruição em massa, terrorismo e ataques cibernéticos. Conforme a missão da OTAN é razoável. As transferências de armas cobrem o fornecimento de armas militares através de vendas, ajuda, presentes e aqueles feitos através de licenças de fabricação. Os dados abrangem as principais armas convencionais, como aeronaves, veículos blindados, artilharia, sistemas de radar, mísseis e navios projetados para uso militar. Excluem-se as transferências de outros equipamentos militares, como armas ligeiras e de pequeno calibre, camiões, pequenas peças de artilharia, munições, equipamento de apoio, transferências de tecnologia e outros serviços. Os valores são Valores Indicativos da Tendência do SIPRI (TIVs) expressos em US $ m. A preços constantes (1990). Um '0' indica que o valor das entregas que são menores que .5 milhões US.

```python
membros=['Belgium',"Bulgaria","Canada","Croatia" "Czech Republic", "Denmark","Estonia","France",
         "Germany", "Greece","Hungary","Iceland", "Italy","Latvia","Lithuania","Netherlands",
         "Norway","Portugal","Romania","Slovak Republic","Spain","Turkey","United Kingdom",
         "United States"]
lista_ex=[]
for a in Exp_Armas.index:
    if a in membros:
        lista_ex.append(True)
    else:
        lista_ex.append(False)
        
Exp_Armas['OTAN'] = lista_ex
Exp_Armas
```
